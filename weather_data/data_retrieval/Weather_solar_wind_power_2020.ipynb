```python
points = [[54.2, 9],[53, 8.4],[53.55, 12.10],[52.3,12], [51,11], [51.15,6.5], [49.7,8],[49.2,11], [48.5,9.3], [51,13], [51.5,11.3],[50.4,9.4]]
def wind_from_points(points, sp, Regions, weights=None):
    seznam = []
    datum = mv.grib_get(sp, ['dataDate','stepRange']) 
    
    for i in points:
        seznam.append(mv.nearest_gridpoint(sp, i))
        seznam[len(seznam)-1].insert(0,int(datum[0][0]))
        seznam[len(seznam)-1].insert(1, Regions[len(seznam) - 1])
    return seznam


def solar_from_points(points, sun, Regions, weights= None):
    seznam = []
    datum = mv.grib_get(sun, ['dataDate','stepRange']) 
    
        
    for i in points:
        seznam.append(mv.nearest_gridpoint(sun, i))
        for j in range(23):
            seznam[len(seznam) - 1][j] = seznam[len(seznam) - 1][j+1] - seznam[len(seznam) - 1][j]
            if j == 22:
                seznam[len(seznam)-1].pop()
                seznam[len(seznam)-1].insert(0,0.0)
                seznam[len(seznam)-1].insert(0,int(datum[0][0]))
                seznam[len(seznam)-1].insert(1, Regions[len(seznam) - 1])
    
    return seznam
    
```

```python
def all_data_wind(info, points, wg2, Regions, ure):
    frames = []
    for datum in datumi2(info):
        u10 = wg2.select(shortName = '10u', dataDate= int(datum))
        v10 = wg2.select(shortName = '10v', dataDate= int(datum))
        sp100 = mv.sqrt(u10*u10 + v10*v10)
        df = pd.DataFrame(wind_from_points(points, sp100, Regions), columns = (["Date", "Region"] + ure))
        frames.append(df)
        
  
    result = pd.concat(frames)
    return result 
```

```python
def all_data_sun(info, points, wg2, Regions, ure):
    frames = []
    for datum in datumi2(info):
        sun = wg2.select(shortName = 'ssrd', dataDate= str(datum))
        df = pd.DataFrame(solar_from_points(points, sun, Regions), columns = (["Date", "Region"] + ure))
        frames.append(df)
        
  
    result = pd.concat(frames)
    return result    
```
