```python
def all_data_wind(info, points, wg, Regions, ure):
    frames = []
    for datum in datumi2(info):
        u10 = wg.select(shortName = '10u', dataDate= int(datum))
        v10 = wg.select(shortName = '10v', dataDate= int(datum))
        sp100 = mv.sqrt(u10*u10 + v10*v10)
        df = pd.DataFrame(wind_from_points(points, sp100, Regions), columns = (["Date", "Region"] + ure))
        frames.append(df)
        
  
    result = pd.concat(frames)
    return result 
```
